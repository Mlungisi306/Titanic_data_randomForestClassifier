```python
import pandas as pd
import numpy as np

#importing the csv files
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')

titanic_data= pd.read_csv('TitanicData-csv.csv')
titanic_data
```

```python
#Performing data cleaning

titanic_data=titanic_data.drop(['PassengerId','Name'],axis=1)
titanic_data['Sex']=titanic_data['Sex'].replace(to_replace=['male','female'], value=[1,0])
titanic_data['Age']=titanic_data['Age'].fillna(titanic_data['Age'].mean())
titanic_data['Survived']=titanic_data['Survived'].replace(to_replace=np.nan, value=-1)
titanic_data['Survived']=titanic_data['Survived'].astype(int)
titanic_data['Embarked']=titanic_data['Embarked'].replace(to_replace=['S','Q','C'], value=[3,2,1])
titanic_data
```

```python
for x in titanic_data['Age']:
    if (x>0.0) & (x<=17.0):
        titanic_data['Age']=titanic_data['Age'].replace(to_replace=x,value='child') # 1= child
    elif (x>17.0) & (x<=35.0):
        titanic_data['Age']=titanic_data['Age'].replace(to_replace=x,value='young adult') # 2= young adult
    elif (x>35.0) & (x<=59):
        titanic_data['Age']=titanic_data['Age'].replace(to_replace=x,value='adult') # 3= adult
    else:
        titanic_data['Age']=titanic_data['Age'].replace(to_replace=x,value='senior citizen') # 4=senior citizen

titanic_data
```

```python
titanic_data['Age']=titanic_data['Age'].replace(to_replace=['child','young adult','adult','senior citizen'],value=[1,2,3,4])
titanic_data
```

```python
# Then performing data Pre-processing
y=titanic_data.iloc[:,0].values
x=titanic_data.iloc[:,1:].values
x
```

```python
x_train=x[:891]
y_train=y[:891]
x_test=x[891:]
y_test=y[891:]
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
# Performing a Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

RF= RandomForestClassifier(n_estimators=100, max_depth=5, random_state=7)
estim= RF.fit(x_train,y_train)
estim
```

```python

from sklearn.model_selection import cross_val_score
score=cross_val_score(estim,x_train,y_train,cv=8)
score.mean()
```

```python
# now making predictions against the test data
y_predict=RF.predict(x_test)
y_predict
```

```python
# creating a new dataframe for submittig the predictions to a kaggle competition project
predictions= pd.read_csv('test-csv.csv')
predictions.insert(1,'Survived',y_predict,True)
predictions.to_csv('RFPredictions.csv',index=False)
predictions
```
